```python
# (1) Create a string
myChars = "Good Morning"

# (2) Convert string to list
myChars_list = list(myChars)

# (3) Convert list to tuple
myChars_tuple = tuple(myChars_list)

# (4) Print all variables
print("String :", myChars)
print("List   :", myChars_list)
print("Tuple  :", myChars_tuple)

# (5) Print the type of each variable
print("Type of myChars      :", type(myChars))
print("Type of myChars_list :", type(myChars_list))
print("Type of myChars_tuple:", type(myChars_tuple))
```

```python
# Creating a nested dictionary

mydict1 = {
    'Ann': {'year': '5', 'score': 87},
    'Tim': {'year': '6', 'score': 90}
}

print('mydict1:', mydict1)

mydict2 = dict([
    ('Ann', ('Year 5', 87)),
    ('Tim', ('Year 6', 90))
])

print('mydict2', mydict2)
```

```python
mydict2['Lila'] = ('Year 6', 89)
mydict2['Tim'] = ('Year 6', 91)

del mydict2['Ann']

print(mydict2)
```

```python
# (1) Create an empty dictionary
fruit_colours = {}

# (2) Add fruit-colour pairs
fruit_colours['Banana'] = 'Yellow'
fruit_colours['Apple'] = 'Red'
fruit_colours['Grapes'] = 'Green'

# (3) Print the dictionary
print("Dictionary:", fruit_colours)

# (4) Print the colour of Apple
print("Apple colour:", fruit_colours['Apple'])

# (5) Update Grapes colour
fruit_colours['Grapes'] = 'Purple'

# (6) Remove Banana
del fruit_colours['Banana']

# (7) Print final dictionary
print("Final Dictionary:", fruit_colours)

# (8) Check data type
print(type(fruit_colours))
```
